## Comoving neighbors to... the Sun?!?

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from astropy import units as u
```

```python
data = Table.read('../data/comovingsun-result-wdist.fits')
```

```python
data2 = Table.read('../data/comovingsun-2mass.csv')
data['2MASS'] = data2['2MASS']
data['Jmag'] = data2['Jmag']
data['e_Jmag'] = data2['e_Jmag']
data['Hmag'] = data2['Hmag']
data['e_Hmag'] = data2['e_Hmag']
data['Kmag'] = data2['Kmag']
data['e_Kmag'] = data2['e_Kmag']
```

```python
data['abs_gmag'] = data['phot_g_mean_mag'] - \
                            5.*(np.log10(data['r_est']) - 1.)
data['abs_gmag'].unit = u.mag
```

```python
plt.scatter(data['bp_rp'], data['abs_gmag'], c='k', alpha=0.7, s=40)
plt.scatter([0.85], [4.83], c='r', marker='*', s=120)
plt.xlabel('Gaia B-R', fontsize=20)
plt.ylabel('Absolute G mag', fontsize=20)
plt.ylim([17,-7])
plt.xlim([-0.8,4.2])
```

```python
plt.hist(data['r_est']/1000., bins=20)
plt.xlabel('Distance (kpc)', fontsize=16)
```
